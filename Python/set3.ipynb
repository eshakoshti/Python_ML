```python
### Question 4
# Write Python code that prints a random number between 1 and 100 and check even odd

import random
random_no = random.randint(1,100)
print("random_no",random_no)

if random_no % 2== 0:
    print(random_no,"is Even Number")
else:
    print(random_no,"is Odd Number")

```
