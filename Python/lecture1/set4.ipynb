```python
# Question 1

# Given: s = "python programming"

# Task: Convert only the first word to uppercase using slicing and string methods → "PYTHON programming"

s="python programing"
print("Uppercase:",s[:6].upper(),s[6:])
```

```python
# Question 2

#  Given: s = " Hello World "

#  Task: Remove spaces on both sides and replace the space between words with - → "Hello-World"
s= "  Hello World  "
print("Remove Space:",s.strip().replace(' ','-'))
```

```python
# Question 3

#  Given: s = "banana"

#  Task: Replace only the last occurrence of "a" with "A" without using loops.
s="banana"
print(s[::-1].replace('a','A',1)[::-1])
```

```python
# Question 4

#  Given: s = "DataScience"

#  Task: Print characters from index 2 to 7 (inclusive of start, exclusive of end) using slicing.

#  Expected Output → "taScie"
s = "DataScience"
print(s[2:8])

```

```python
# Question 5

#  Given: s = "helloHELLO"

#  Task: Swap case of each character and remove the last 3 characters using slicing.

#  Expected Output → "HELLOhe"

s = "helloHELLO"
print(s.swapcase()[:-3])
```

```python
# Question 6

#  Given: s = "I love python programming"

#  Task: Using slicing, reverse only the substring "python" while keeping the rest same.

#  Expected Output → "I love nohtyp programming"

s = "I love python programming"
result = s[:7] + s[7:13][::-1] + s[13:]
print(result)
```

```python
#  Question 7

#  Given: s = "mississippi"

#  Task: Count how many times "ss" appears in the string using a string method.

s = "mississippi"
print(s.count('ss'))

```

```python
#  Question 8

#  Given: s = "machinelearning"

#  Task: Insert a space after "machine" using slicing.

#  Expected Output → "machine learning"

s = "machinelearning"
print(s[:7],(""),s[7:])

```

```python
#  Question 9

#  Given: s = "Welcome123"

#  Task: Check if the string contains only alphabetic characters after removing digits and print the result (True/False).

s = "Welcome123"
s_without_digits = ''.join(char for char in s if not char.isdigit())
print(s_without_digits)

# Check if alphabetic
print(s_without_digits.isalpha())
```
