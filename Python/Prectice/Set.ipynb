```python
myset={'Hindi','Gujrati','English'}
print(myset)
```

```python
# Add  Item

myset={'Hindi','Gujrati','English'}
myset.add('Panjabi')
print(myset)
```

```python
#  Add item myset into Language
Language={'Hindi','Gujrati','English','Panjabi'}
myset={1,2,3,4}
myset.update(Language)
print(myset)
```

```python
myset={'Hindi','Gujrati','English','panjabi'}
myset.remove('English')
print(myset)
```

```python
# Using Loop  
myset={'Hindi','Gujrati','English'}
for s in myset:
    print(s)   
```

```python
#  using Union method
myset={'Hindi','Gujrati','English'}
myset1={1,2,3}
result=myset.union(myset1)
print(result)
```
