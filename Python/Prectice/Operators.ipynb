```python
#  Assignment Operators

n=2
n **= 3
print(n)
```

```python
# Logical operator And
n=10
print(n > 8 and n < 20)
print(n > 20 and n < 5 )
    
```

```python
# Logical operator or
n=15
print(n > 30 or n < 10)
print(n > 8 or n < 10)

```

```python
# Identify Operator
# is check object in memory same or not
# == check the variable same or not 

a=['red','green','yellow','pink']
b=['red','green','yellow','pink']
print(a is b)
print(a is not b)
print(a == b)

```

```python
# Membership operator

color=['red','green','yellow','pink']
print("yellow" in color)
print("pink" not in   color)
```

```python
# Comparison operator
x=2
y=5 
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)

```

```python
# Bitwise operator

print(4 & 6)
print(4 | 6)
print(~4)
print(4 ^ 6)
print(4 << 2)
print(4 >> 2)
```
