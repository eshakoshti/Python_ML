```python
# Create List

mylist=['vivo','samsaung','oppo']
print(mylist)
```

```python
# Add the one Item:

mylist=['vivo','samsaung','oppo']
mylist.append("apple")
print("After Appending list=",mylist)
```

```python
# Remove  the list

mylist=['vivo','samsaung','oppo','apple']
mylist.remove("oppo")
print(mylist)
```

```python
# Access the list
mylist=['vivo','samsaung','apple']
print(mylist[0])
print(mylist[1])
```

```python
mylist=['vivo','samsaung','apple','realme','redmi']
print(mylist[1:4])

```

```python
# change list Item

mylist=['vivo','samsaung','apple']
mylist[1]=['Realme']
print(mylist)
```

```python
#  Using Loop  List

mylist=['vivo','samsaung','apple']
for i in mylist:
    print(i)
```

```python
# Using while loop list

mylist=['vivo','samsaung','apple']
i=0
mylist_len=len(mylist)
while i<mylist_len:
    print(mylist[i])
    i+=1
  
```

```python
# Using Range and Len()

mylist=['vivo','samsaung','apple','Realme']
for i in range(len(mylist)):
    print(mylist[i])
```

```python
# using while and len()

mylist=['vivo','samsaung','apple']
i=0
while i<len(mylist):
    print(mylist[i])
    i+=1 
```

```python
# Sort List
mylist=['vivo','Samsaung','apple','Realme','oppo']
mylist.sort()
print(mylist)
```

```python

mylist=['vivo','Samsaung','Apple','Realme','oppo']
mylist.sort(key=str.lower)
print(mylist)
```

```python
# reverse list

mylist=['vivo','apple','realme','oppo']
mylist.reverse()
print(mylist)
```

```python
# join two list
mylist=['vivo','apple','realme','oppo']
mylist1=[1,2,3,4]
result=mylist + mylist1
print(result)
```

```python
# using extend() method

mylist=['vivo','apple','realme','oppo']
mylist1=[1,2,3]
mylist1.extend(mylist)
print(mylist1)
```

```python
# copy() method

mylist1=[1,2,3,4,5]
result=mylist1.copy()
print(result)
```

```python
# count() method

mylist1=[1,2,3,4,5,3,4,2,3,5,5]
no=mylist1.count(5)
print(no)

```

```python
# index() method

mylist1=[1,2,3,4,5]
mylist1.index(4)
```

```python
# Using pop() method

mylist1=[1,2,3,4,5,6]
mylist1.pop()
print(mylist1)

```

```python
 # using clear()

mylist=[1,2,3,4,5,5,6]
mylist.clear()
print(mylist)


```
