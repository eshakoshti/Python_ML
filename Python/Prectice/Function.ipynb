```python
#  Keyword Argument

def keyarg(fn,ln):
    print("First Name:",fn)
    print("First: ",fn +"lastName:",ln)
keyarg(ln="Patel",fn="Savi ")
```

```python
# position Argument

def employee(name,city):
    print("Name:",name)
    print("City:",city)
employee("Nidhi","Ahemdabad")
```

```python
#  Variable Length Argument and keyword Length Argument
def mul(*args,**kwargs):
    mul=1
    for i in args:
        mul=i*mul
    print("Multiplication :",mul)
    print("Hello")
    
mul(10,20)
mul(30)
mul(10,20,30,day="Monday",work="good") 
```

```python
# kwargs with regular Argument

def employee(name,age,**args):
    print("Name:",name)
    print("Age:",age)
    print(args)
employee("Pooja",25,city="Surat",email="pooj@gmail.com")
```

```python
# Local Scop

def myfun():
    x='welcome'
    def myvalue():
        print(x)
    myvalue()
myfun()
```

```python
# global Scop
x= 'Hello'
def myfun():
    print(x)
    def myvalue():
       print("Global value:",x)
    myvalue()
myfun()
```

```python
# global keyword
def myfun():
    global x
    x=9
    print("Access the local variable using global keyword")
myfun()  
```

```python
# Lambda Function
x = lambda a,b : a+b
print(x(5,5))
```

```python
# Filter using
no=[2,4,3,5,8,7,9,10,34,35,80]
result=list(filter(lambda x:x%2==0,no))
print(result)
```

```python
# Using Map()
no=[1,2,3,4,5,6]
result=list(map(lambda x: x * x,no))
print(result)
```

```python
# Recursive function

def factorial(n):
    if n==1:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(5))
```
