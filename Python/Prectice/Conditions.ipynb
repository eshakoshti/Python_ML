```python
color=3
match color:
    case 1:
        print('Green')
    case 2:
        print('yellow')
    case 3:
        print('Blue')
    case 4|5|6:
        print('Pink')
    case 7:
        print('Red')
    case _:
        print('No match')
```

```python
#   While loop

str='welcome'
len_value=len(str)
i=0
while i<len_value:
    print(str[i],end="")
    i+=1

```

```python
#  Break ststement

str='Hello Everyone Good morning!'
for i in str:
    if i=='G':
        break
    print(i,end="") 
```

```python
#  Break Stastement using while
i=0
str='Welcome to my Page'
while i < len(str):
    if i==10:
        break
    print(str[i],end="")
    i+=1
```

```python
#  Continue Statement
i=0
while i < 10:
    i+=1
    if i==5:
        continue
    print(i,end=" ")
```
